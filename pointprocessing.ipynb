```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

image = Image.open("assets/sample.jpg")
image_array = np.array(image)
imgplot = plt.imshow(image)
```

```python
xlim = image_array.shape[0]
ylim = image_array.shape[1]
zlim = image_array.shape[2]
image_array.shape
```

```python
# Brightness
image_array_2 = np.zeros(image_array.shape, image_array.dtype)
for x in range(xlim):
    for y in range(ylim):
        for z in range(zlim):
            # Brightness adjustment
            image_array_2[x][y][z] = np.clip(1.10 * image_array[x][y][z] + 50, 0, 255)
            # Contrast enhancement
            #image_array_2[x][y][z] = np.clip(1.75 * (image_array[x][y][z] - 128) + 128, 0, 255)

im = Image.fromarray(image_array_2)
im.save("assets/sample_2.jpg");

image2 = Image.open("assets/sample_2.jpg")

f,axarr = plt.subplots(1,2,figsize=(30,30))
axarr[0].imshow(image)
axarr[1].imshow(image2)
```
