```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt

image = Image.open('assets/sample.jpg')
image_array = np.array(image)
new_size = (image_array.shape[1] // 2, image_array.shape[0] // 2)
resized_image = image.resize(new_size)
resized_image.save('assets/resized_sample.jpg')

f,axarr = plt.subplots(1,2,figsize=(15,15))
axarr[0].imshow(image)
axarr[1].imshow(resized_image)
```

```python
angle = 45
rotated_image = image.rotate(angle, expand=True)
rotated_image.save('assets/rotated_sample.jpg')

f,axarr = plt.subplots(1,2,figsize=(15,15))
axarr[0].imshow(image)
axarr[1].imshow(rotated_image)
```

```python
x_offset = 100
y_offset = 50

# Apply the translation
translated_image = image.transform(
    image.size, 
    Image.AFFINE, 
    (1, 0, x_offset, 0, 1, y_offset)
)

translated_image.save('assets/translated_sample.jpg')

f,axarr = plt.subplots(1,2,figsize=(15,15))
axarr[0].imshow(image)
axarr[1].imshow(translated_image)
```
